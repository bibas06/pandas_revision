```python
def get_city_match_count(city):
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count("Bangalore")
```

```python
def city_date(city,date):
    mask1=data['city']==city
    mask2=data['date']>date
    return data[mask1 & mask2].shape[0]
```

```python
city_date("Hyderabad","2010-01-01")
```
